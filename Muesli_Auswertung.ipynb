```python
import pandas as pd
```

```python
df_orders = pd.read_csv('data/5 - Muesli Project raw data - Orders.csv', skiprows = 1)



cols = df_orders.columns.tolist()

cols = [col.replace(' ', '_') for col in cols]

df_orders.columns = cols





```

```python
df_orders['Order_Date'] = pd.to_datetime(df_orders['Order_Date'], dayfirst=True, format="%d/%m/%y", infer_datetime_format=True)
df_orders.drop(["Index"], axis=1, inplace=True)
df_orders.head(30)

```

```python
df_orders.info()
```

```python
df_campaign_data = pd.read_csv('data/5 - Muesli Project raw data - Campaign Data.csv')
df_campaign_data.head()

cols1 = df_campaign_data.columns.tolist()

cols1 = [col.replace(' ', '_') for col in cols1]

df_campaign_data.columns = cols1

df_campaign_data['Arrival_Scan_Date'] = pd.to_datetime(df_campaign_data['Arrival_Scan_Date'], dayfirst=True, format="%d/%m/%y", infer_datetime_format=True)

df_campaign_data.head(20)


```

```python
df_order_process_data = pd.read_csv('data/5 - Muesli Project raw data - Order Process Data.csv')
df_order_process_data.head()

cols2 = df_order_process_data.columns.tolist()

cols2 = [col.replace(' ', '_') for col in cols2]

df_order_process_data.columns = cols2

df_order_process_data['Order_Date'] = pd.to_datetime(df_order_process_data['Order_Date'], infer_datetime_format=True)
df_order_process_data['On_Truck_Scan_Date'] = pd.to_datetime(df_order_process_data['On_Truck_Scan_Date'])

df_order_process_data.head()
```

```python
df_intern_data_study = pd.read_csv('data/5 - Muesli Project raw data - InternData Study.csv')
df_intern_data_study.head()

cols3 = df_intern_data_study.columns.tolist()

cols3 = [col.replace(' ', '_') for col in cols3]

df_intern_data_study.columns = cols3

df_intern_data_study['Ready_to_Ship_Date'] = pd.to_datetime(df_intern_data_study['Ready_to_Ship_Date'])
df_intern_data_study['Pickup_Date'] = pd.to_datetime(df_intern_data_study['Pickup_Date'])

df_intern_data_study.head(20)
```
